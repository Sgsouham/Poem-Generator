```python
import gpt_2_simple as gpt2
from datetime import datetime
```

```python
sess = gpt2.start_tf_sess()
gpt2.load_gpt2(sess, run_name='run1')
```

```python
def gen_poem(text):
  print(gpt2.generate(sess,
              length=250,
              temperature=0.7,
              prefix=text,
              nsamples=1,
              batch_size=1
              ))
```

```python
print("Enter a poem - ")
text = input()
gen_poem(text)
```
